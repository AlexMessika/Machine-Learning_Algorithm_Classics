## faire un bon graphique ? 

Quels sont les éléments indispensables à un bon graphique ? 

- un titre
- légende
- **axes** avec DES UNITES 
- couleur lisible : jaune sur blanc => 0 au module
- graphiques pas surchargés : il vaut mieux faire plusieurs graphiques qui présentent chacun un aspect du phénomène plutôt que de tout avoir sur un seul graphique de manière illisible
- Type du graphique judicieux avec les données dont on dispose
- Un graphique doit avoir une raison d'être : il faut qu'il permette de soulever des questions ou d'y repondre : de soutenir une analyse. 


## Analyser un graphique

Quand on analyse un graphique on dit pas juste "oh on voit que le prix augmente avec la taille d'une maison", car merci on le voit. Il faut aller plus loin. Par exemple en émettant des hypothèse ou donner des explications à propos du phénomène observer. Faire ressortir les informations importantes ou les choses qui ne sont pas évidentes au premier coup d'oeil






```python
#un premier graphique qui ne respecte quasiement aucun critère avec matplotlib

import matplotlib.pyplot as plt
import numpy as np 


plt.plot([0.5, 1, 1.5, 2, 2.5], [2000, 2200, 2150, 3000, 2400])
```

```python
# un peu mieux : on ajoute un titre

plt.title("Salaire en fonction experience")
plt.plot([0.5, 1, 1.5, 2, 2.5], [2000, 2200, 2150, 3000, 2400])
```

```python
# on approche du minimum vital : axes avec des unités
plt.title("Salaire en fonction experience")
plt.plot([0.5, 1, 1.5, 2, 2.5], [2000, 2200, 2150, 3000, 2400]) # si j'ai un plot yellow c'est 0!!!!!!!!!!!!!
plt.xlabel("Experience (année)")
plt.ylabel("Salaire mensuel (€)")
```

```python
#options pour modifier les tailles de la figure
plt.rcParams['figure.figsize'] = (10, 6)
```

## les types de graphiques

Il existe de nombreux types de graphiques différents :
- lignes
- nuages de points (scatter plot)
- bar
- bar empilées (stacked bar)
- camember (attention pas toujours terrible)

On va voir ici l'intérêt des nuages de points

```python
# prenons des points généré aléatoirement
# et tentons comme avant. 

v1 = np.random.randn(40)
v2 = np.random.randn(40)


plt.plot(v1, v2)
```

pas terrible... 
et si on faisait un nuage de point plutôt ? 

```python
plt.scatter(v1, v2) #nuage de point
```

Les diagrames en bar sont souvent très pratiques pour compter des catégories ou des populations. 


```python
plt.bar(['12-25', '25-50', '50+'],   height = np.array([20, 10, -5]))
```

```python
# est ce que je peux avoir plusieurs graphiques côte à côte ? 
# la réponse est oui, avec subplot
```

```python
fig, subplots = plt.subplots(2, 2) # jeveux des graphiques sur 2 lignes et 2 colonnes
subplots[0, 0].plot([1, 2, 3], [2, 2, 5])
subplots[0, 1].scatter([1, 2, 3], [-1, 6, 2])
subplots[1, 0].bar(['mineur', 'majeur'],  height=[20, -5])
subplots[1, 1].plot([1, 2, 3])
```
