pas content, le chat a une drole de tte 
c'est parce que imread de opencv ne charge pas l'image dans l'ordre RGB mais il le fait dans BGR

On peut convertir

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
type(img2)
```

```python
plt.imshow(img2)
```

passer de la couleur  une image en niveau de gris : moyenne des couleur pixel par pixel

```python
img_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
img_gray.shape
```

```python
img_gray2 = img_gray.reshape(480, 640, 1)
img_gray2.shape
```

```python
plt.imshow(img_gray2)
```

```python
plt.imshow(img_gray2, cmap="gray")
```

La conversion vers les niveaux de gris = moyenne 
et du coup on peut directement en numpy

```python
img_gray = img.mean(axis=2)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```
