```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Parameters
n_classes = 3
plot_colors = "ryb"
plot_step = 0.02

# Load data
iris = load_iris()


   
```

```python
# Test Fitting
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = None, None, None, None
X_train, X_test, y_train, y_test= train_test_split(X, y, train_size=100, test_size=50, random_state=0)

from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion="entropy",max_depth=3)

clf = clf.fit(X_train, y_train)
test=clf.predict(X_test, clf )

# Test accuracy
moyenne = y_test
cpt = 0
for i in range (0,y_test.size-1):
  if moyenne[i] - test[i] != 0:
    cpt+=1
cpt = 100-(cpt/y_test.size)*100

print("confusion:\n",confusion_matrix(y_test,test),"\n")
print("accuracy:\n",accuracy_score(y_test,test),"\n")
print("classReport:\n",classification_report(y_test,test),"\n")


```

```python
plt.figure(figsize=(15,15))
plot_tree(clf)
```
